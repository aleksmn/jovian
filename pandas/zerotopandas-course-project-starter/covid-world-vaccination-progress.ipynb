## Data Preparation and Cleaning



```python
import pandas as pd
```

```python
world_df = pd.read_csv('covid-world-vaccination-progress/country_vaccinations.csv')
```

```python
world_df.sample(10)
```

```python
world_df.shape
```

```python
world_df.info();
```

```python
world_df.date = pd.to_datetime(world_df['date'])
world_df.date
```

```python
world_df['year'] = pd.DatetimeIndex(world_df.date).year
world_df['month'] = pd.DatetimeIndex(world_df.date).month
world_df['day'] = pd.DatetimeIndex(world_df.date).day
```

Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
world_df.describe()
```

```python
# List of countries in dataset
world_df['country'].unique()
```

```python
russia_df = world_df[world_df.country == 'Russia']
usa_df = world_df[world_df.country == 'United States']
china_df = world_df[world_df.country == 'China']
```

## Compare the dynamic of vaccinations in different countries

```python
plt.plot(russia_df.date, russia_df.daily_vaccinations_per_million)
plt.plot(usa_df.date, usa_df.daily_vaccinations_per_million)
plt.plot(china_df.date, china_df.daily_vaccinations_per_million)

plt.title("Daily vaccinations from COVID-19")
plt.legend(["Russia", "United States", 'China'])
plt.xlabel("Date")
plt.ylabel("Vaccinated per million");
```

## Total vaccination per hundred of people

```python
plt.plot(russia_df.date, russia_df.total_vaccinations_per_hundred)
plt.plot(usa_df.date, usa_df.total_vaccinations_per_hundred)

plt.title("Total vaccinations per hundred people")
plt.legend(["Russia", "United States"]);
```

Compare vaccinations in 2020 and 2021:

```python
years = world_df.groupby('year').sum().total_vaccinations
```

```python
for k, v in years.items():
    print(f"In year {k} vaccinated {int(v//1000000)} million people.")
```

Explore dynamic of vaccination is year 2021.

```python
months_df = world_df[world_df.year == 2021].groupby('month')
months_df.mean()
```

```python
plt.bar(months_df.mean().index, months_df.mean().daily_vaccinations)


plt.title("World daily vaccination by month")
plt.xlabel('Month of 2021')
plt.ylabel('Mean daily vaccination');
```

```python
plt.bar(months_df.mean().index, months_df.mean().people_vaccinated)


plt.title("People vaccinated")
plt.xlabel('Month of 2021')
plt.ylabel('Mean daily people vaccinated');
```

## People vaccinated per hundred

```python
vaccinated_per_hundred_df = world_df.groupby('country').max()['people_vaccinated_per_hundred']
vaccinated_per_hundred_df = vaccinated_per_hundred_df[['United States', 'Germany',  'Turkey', 'Russia', 'India']]
```

```python
sns.barplot(x=vaccinated_per_hundred_df.index, y=vaccinated_per_hundred_df)

plt.title('People vaccinated per hundred')
plt.ylabel('vaccinated per hundred');
```

## People vaccinated

```python
people_vaccinated_df = world_df.groupby('country').max()['people_vaccinated']
people_vaccinated_df = people_vaccinated_df[['United States', 'Germany',  'Turkey', 'Russia', 'India']]
```

```python
sns.barplot(x=people_vaccinated_df.index, y=people_vaccinated_df // 1000000);
plt.title('ToPeople vaccinated')
plt.ylabel('vaccinated people, mln');
```
